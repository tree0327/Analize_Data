```python
import pandas as pd

# 예시 상품 판매 데이터
sales_data = {
    'OrderID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Product': ['Laptop', 'Mouse', 'Laptop', 'Monitor', 'Mouse', 'Keyboard', 'Laptop', 'Mouse', 'Monitor', 'Keyboard'],
    'Quantity': [1, 2, 1, 1, 2, 1, 1, 3, 1, 1],
    'Price': [1200, 25, 1200, 300, 25, 100, 1200, 25, 300, 100],
    'CustomerID': [1001, 1002, 1001, 1003, 1004, 1002, 1001, 1003, 1005, 1002],
    'OrderDate': ['2024-01-05', '2024-01-06', '2024-01-07', '2024-01-10', '2024-01-12', '2024-01-15', '2024-01-18', '2024-01-20', '2024-01-22', '2024-01-25']
}
df_sales = pd.DataFrame(sales_data)
df_sales
```

# 1번

```python
# 총 구매금액으로 새로운 컬럼 만들기
df_sales['Total'] = df_sales['Quantity'] * df_sales['Price']

# 고객별 총 구매금액 계산하기
customer_total = df_sales.groupby('CustomerID')['Total'].sum()

# 결과 출력
customer_total
```

# 2번 - Laptop을 구매한 고객들의 평균 구매 금액

```python
# 1. 랩탑 구매자 데이터 구하기
laptop_sales = df_sales[df_sales['Product'] == 'Laptop']
# laptop_sales

# 2. 랩탑 구매자의 총 구매금액의 평균
laptop_avg_purchase = laptop_sales['Total'].mean()
laptop_avg_purchase
```

# 3번 - 가장 많이판매된 상품의 이름과 판매된 총 수량

```python
# 상품별 판매수량 합계 계산 (groupby 사용 reset_index 사용 안함)
product_sales = df_sales.groupby('Product')['Quantity'].sum()
# product_sales
# idxmax(): 가장 큰 값을 가진 인덱스 반환 / max(): 가장 큰 값을 반환
best_product = product_sales.idxmax()
best_product_qt = product_sales.max()
print(f"가장 많이 판매된 상품: {best_product} ({best_product_qt}개)")
```

# 4번 - OrderDate가 2024년 1월 10일 이후인 주문만 필터링하여 출력

```python
df_sales
# 1. 대소비교를 위해 OrderDate 컬럼의 값을 datetime 형식으로 반환받음
df_sales['OrderDate'] = pd.to_datetime(df_sales['OrderDate'])
# df_sales.info()

# 2 OrderDate > '2024-01-10'인 데이터 출력
filtered_sales = df_sales[df_sales['OrderDate'] > '2024-01-10']
filtered_sales
```

# 5번 - Product별 판매량과 총 판매 금액을 계산한 후, 이를 총. ㅏㄴ매금액 기준으로 내림차순으로 정렬

```python
# 상품별 판매량과 총 판매금액(agg함수 사용)
product_stats = df_sales.groupby('Product').agg({'Quantity': 'sum', 'Total': 'sum'})
product_stats

# 총 판매금액 기준으로 내림차순 정렬
sorted_stats = product_stats.sort_values(by='Total', ascending=False)
sorted_stats
```
