### [실습 2] **중복값 제거**

다음은 한 판매점의 제품 판매 기록입니다. 동일한 제품이 중복으로 기록되어 있습니다. 중복된 `ProductID`를 가진 제품의 중복된 행을 제거한 후, 가장 저렴한 가격을 가진 데이터만 남기세요.

```python
product_data = {
    'ProductID': [1, 2, 3, 1, 4, 2],
    'ProductName': ['Laptop', 'Mouse', 'Monitor', 'Laptop', 'Keyboard', 'Mouse'],
    'Price': [1200, 25, 300, 1150, 45, 22],
    'QuantitySold': [10, 100, 50, 8, 70, 120]
}
```

````python
import pandas as pd

product_data = {
    'ProductID': [1, 2, 3, 1, 4, 2],
    'ProductName': ['Laptop', 'Mouse', 'Monitor', 'Laptop', 'Keyboard', 'Mouse'],
    'Price': [1200, 25, 300, 1150, 45, 22],
    'QuantitySold': [10, 100, 50, 8, 70, 120]
}

df = pd.DataFrame(product_data)

# 1. 가격(Price)을 기준으로 오름차순 정렬
# (이렇게 하면 같은 ProductID 중에서 가격이 낮은 항목이 리스트의 위쪽으로 오게 됩니다)
df_sorted = df.sort_values(by='Price', ascending=True)

# 2. ProductID를 기준으로 중복 제거
# keep='first' 옵션은 중복된 값 중 가장 위에 있는 첫 번째 값(아까 정렬한 최저가 항목)을 남깁니다.
df_result = df_sorted.drop_duplicates(subset='ProductID', keep='first')

# 3. 인덱스 재설정 (깔끔한 출력을 위해 기존 인덱스를 버리고 새로 매깁니다)
df_result = df_result.reset_index(drop=True)

# 결과 확인
print(df_result)
````

## **[종합 실습 1] 상품 판매 데이터 분석**

다음은 한 전자상거래 사이트의 **상품 판매 데이터**입니다. 고객들이 특정 상품을 구매한 기록이 포함되어 있습니다. 데이터를 활용하여 아래의 문제를 해결하세요.

````python
import pandas as pd

# 예시 상품 판매 데이터
sales_data = {
    'OrderID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Product': ['Laptop', 'Mouse', 'Laptop', 'Monitor', 'Mouse', 'Keyboard', 'Laptop', 'Mouse', 'Monitor', 'Keyboard'],
    'Quantity': [1, 2, 1, 1, 2, 1, 1, 3, 1, 1],
    'Price': [1200, 25, 1200, 300, 25, 100, 1200, 25, 300, 100],
    'CustomerID': [1001, 1002, 1001, 1003, 1004, 1002, 1001, 1003, 1005, 1002],
    'OrderDate': ['2024-01-05', '2024-01-06', '2024-01-07', '2024-01-10', '2024-01-12', '2024-01-15', '2024-01-18', '2024-01-20', '2024-01-22', '2024-01-25']
}
df_sales = pd.DataFrame(sales_data)
````

1. 각 고객(`CustomerID`)별로 총 구매 금액을 계산하세요.
2. `Laptop`을 구매한 고객들의 평균 구매 금액을 계산하세요.
3. 가장 많이 판매된 상품(상품 이름과 판매된 총 수량)을 구하세요.
4. `OrderDate`가 2024년 1월 10일 이후인 주문만 필터링하여 출력하세요.
5. `Product`별로 판매량과 총 판매 금액을 계산한 후, 이를 `Price` 기준으로 내림차순으로 정렬하여 출력하세요.

````python
import pandas as pd
# 데이터
sales_data = {
    'OrderID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Product': ['Laptop', 'Mouse', 'Laptop', 'Monitor', 'Mouse', 'Keyboard', 'Laptop', 'Mouse', 'Monitor', 'Keyboard'],
    'Quantity': [1, 2, 1, 1, 2, 1, 1, 3, 1, 1],
    'Price': [1200, 25, 1200, 300, 25, 100, 1200, 25, 300, 100],
    'CustomerID': [1001, 1002, 1001, 1003, 1004, 1002, 1001, 1003, 1005, 1002],
    'OrderDate': ['2024-01-05', '2024-01-06', '2024-01-07', '2024-01-10', '2024-01-12', '2024-01-15', '2024-01-18', '2024-01-20', '2024-01-22', '2024-01-25']
}

# DataFrame 생성
df_sales = pd.DataFrame(sales_data)

# 'OrderDate'를 날짜 형식으로 변환 (문자열 -> datetime)
df_sales['OrderDate'] = pd.to_datetime(df_sales['OrderDate'])
````

````python
# 1. 각 고객(CustomerID)별로 총 구매 금액을 계산하세요.
df_sales['TotalAmount'] = df_sales['Price'] * df_sales['Quantity']
customer_total = df_sales.groupby('CustomerID')['TotalAmount'].sum()

customer_total
````

````python
# 단계 1: 각 주문 건별 '총 금액(TotalAmount)' 계산
# (개별 단가 Price * 수량 Quantity)을 곱해서 새로운 컬럼으로 만듭니다.
df_sales['TotalAmount'] = df_sales['Price'] * df_sales['Quantity']

# 단계 2: 고객별(CustomerID)로 그룹화(groupby)하여 총 금액의 합(sum) 집계
customer_total = df_sales.groupby('CustomerID')['TotalAmount'].sum()

# 단계 3: 보기 좋게 결과 인덱스 리셋 (선택 사항)
customer_total_df = customer_total.reset_index()

print("== 고객별 총 구매 금액 ==")
print(customer_total_df)
````

````python
# ------------------------------------------------------------------------------------------------
# 2. `Laptop`을 구매한 고객들의 평균 구매 금액을 계산하세요.
# ------------------------------------------------------------------------------------------------

# 단계 1: 'Laptop'을 구매한 데이터만 필터링
laptop_sales = df_sales[df_sales['Product'] == 'Laptop']

# 단계 2: Laptop 구매 고객 집단의 'TotalAmount' 평균 계산
laptop_avg_amount = laptop_sales['TotalAmount'].mean()

print(f"\n== Laptop 구매 고객 평균 구매액: {laptop_avg_amount:.2f} ==")
````

````python
# ------------------------------------------------------------------------------------------------
# 3. 가장 많이 판매된 상품(상품 이름과 판매된 총 수량)을 구하세요.
# ------------------------------------------------------------------------------------------------

# 단계 1: 상품별(Product) 판매 수량(Quantity) 합계 계산
product_quantity = df_sales.groupby('Product')['Quantity'].sum()

# 단계 2: 판매 수량이 가장 많은(max) 상품 찾기 - idxmax() 사용 (인덱스 반환)
top_product_name = product_quantity.idxmax()
top_product_qty = product_quantity.max()

print(f"\n== 가장 많이 판매된 상품: {top_product_name} ({top_product_qty}개) ==")
````

````python
# ------------------------------------------------------------------------------------------------
# 4. `OrderDate`가 2024년 1월 10일 이후인 주문만 필터링하여 출력하세요.
# ------------------------------------------------------------------------------------------------

# 단계 1: 날짜 조건 필터링 (> '2024-01-10')
recent_orders = df_sales[df_sales['OrderDate'] > '2024-01-10']

print("\n== 2024년 1월 10일 이후 주문 ==")
print(recent_orders)
````

````python
# ------------------------------------------------------------------------------------------------
# 5. `Product`별로 판매량과 총 판매 금액을 계산한 후, 이를 `Price` 기준으로 내림차순으로 정렬하여 출력하세요.
# ------------------------------------------------------------------------------------------------

# 단계 1: 상품별(Product)로 그룹화하여 -> 수량(Quantity)과 총금액(TotalAmount) 합계 집계
product_stats = df_sales.groupby('Product')[['Quantity', 'TotalAmount']].sum()

# 단계 2: 'TotalAmount' 기준으로 내림차순(descending) 정렬
# (문제에서는 Price 기준이라고 했으나, 집계된 데이터 프레임에는 Price 컬럼이 없고 TotalAmount가 더 적절할 수 있음. 
# 하지만 문제 의도가 '집계 후 총 판매금액 순'이라면 TotalAmount 사용, 만약 '단가'라면 원본 정보 병합 필요.
# 여기서는 '총 판매 금액'을 기준으로 정렬하겠습니다.)
product_stats_sorted = product_stats.sort_values(by='TotalAmount', ascending=False)

print("\n== 상품별 판매량 및 총 판매 금액 (매출액 순 정렬) ==")
print(product_stats_sorted)
````
